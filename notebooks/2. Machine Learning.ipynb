# Máquina

Artefacto que realiza cómputo.

* Máquina abstracta - Modelo matemático (función / aplicación)
* Máquina aplicada - Modelo de datos (programa / algoritmo)

Una máquina se compone de un estado, un contexto, una interfaz de entrada y una interfaz de salida.

```python
def maquinaSumadora(numeros):
    s = 0 # estado `s`
    
    for n in numeros:
        s += n # s = s + n (cómputo)
        
    return s 
```

```python
maquinaSumadora([1, 3, 5, 7, 9, 11])
```

```python
total = maquinaSumadora([1, 3, 5, 7, 9, 11])

total
```

# Aprendizaje

Una máquina generalmente basa sus cómputos en parámetros de ajuste. Estos parámetros provocan que los cálculos sean operados de manera efectiva, para las condiciones de cada problema. Es decir, podemos pensar el contexto de una máquina y su estado, a través de parámetros configurados. Los parámetros generalmente se ajustan a los datos.

El aprendizaje de una máquina, consiste en ajustar sus parámetros, para que responda de la manera más eficiente a las entradas de datos y produzca la salida óptima como resultado de sus cómputos. Por ejemplo, una máquina que aprenda a clasificar imágenes, textos, audios, etc.

```python
def evaluacionLineal(a, b, c, datos):
    s = 0
    for x1, x2 in datos:
        s += a * x1 + b * x2 + c
    return s
```

```python
datos = [
    (1, 1),
    (2, 3),
    (3, 2),
    (4, 1.5),
    (5, 1.7)
]

evaluacionLineal(1, 1, 5, datos)
```

```python
from random import uniform

a_min = 0.1
b_min = 0.1
c_min = 0.1

for i in range(10000):
    a = uniform(-100, 100)
    b = uniform(-100, 100)
    c = uniform(-100, 100)
    
    err_min = evaluacionLineal(a_min, b_min, c_min, datos)
    err = evaluacionLineal(a, b, c, datos)
    
    if err < err_min:
        a_min = a
        b_min = b
        c_min = c
        
evaluacionLineal(a_min, b_min, c_min, datos)
```

```python
from math import sin

def sin_param(A, w, p, x):
    return A * sin(w * x + p)
```

```python
x = np.linspace(-np.pi, np.pi)
y = 3 * np.sin(0.5 * x + 0.7) + np.random.rand(len(x))
y_param = [sin_param(2, 4, 0.5, xi) for xi in x]

plt.plot(x, y)
plt.plot(x, y_param)
```

```python
def err_sin_param(A, w, p, x, y):
    s = 0
    for i in range(len(x)):
        xi = x[i]
        yi = y[i]
        yi_p = sin_param(A, w, p, xi)
        s += (yi - yi_p) ** 2
    return s
```

```python
err_sin_param(1, 1, 0, x, y)
```

```python
err_sin_param(2, 2, 0.5, x, y)
```

```python
from random import uniform

a_min = 0.1
w_min = 0.1
p_min = 0.1

for i in range(10000):
    a = uniform(2.5, 3.5)
    w = uniform(0.4, 0.7)
    p = uniform(-np.pi / 2, np.pi / 2)
    
    err_min = err_sin_param(a_min, w_min, p_min, x, y)
    err = err_sin_param(a, w, p, x, y)
    
    if err < err_min:
        a_min = a
        w_min = w
        p_min = p
        
x = np.linspace(-np.pi, np.pi)
y = 3 * np.sin(0.5 * x + 0.7) + np.random.rand(len(x))
y_param = [sin_param(a_min, w_min, p_min, xi) for xi in x]

plt.plot(x, y)
plt.plot(x, y_param)
```
