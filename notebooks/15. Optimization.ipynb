```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (x - 3) ** 2 + 5

x = np.linspace(-20, 20, 20)
y = [f(xi) for xi in x]

plt.plot(x, y)
```

```python
xt = 0

h = 0.0001

for i in range(100):
    df = (f(xt + h) - f(xt - h)) / (2 * h)
    xt = xt - f(xt) / df

xt, f(xt)
```
