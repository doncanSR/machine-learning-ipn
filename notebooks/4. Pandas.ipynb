## Series

1. Son vectores o arreglos de numpy en transfondo
2. Es que se conceptualizan como columnas en un dataset
3. Pueden ser nombradas para unirse a dataframes

```python
import pandas as pd # Importaci√≥n de pandas con el alias pd

pd.Series([1, 2, 3, 4])
```

```python
import numpy as np

x = pd.Series([1, 2, 3, 4])
y = x * 3 + np.sin(x)

y
```

```python
pd.Series([1, 2, 3, 4], name="Edades")
```

## DataFrames

1. Son matrices en transfondo
2. Se conceptualizan como tablas (filas y columnas)
3. Podemos agrupar series como columnas

```python
pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [2, 6, 5, 3],
    "C": [False, True, False, False]
})
```

```python
A = pd.Series([1, 2, 3, 4], name="A")
B = A ** 2
B.name = "B"
C = B > 5
C.name = "C"

pd.DataFrame([A, B, C]).T
```

```python
data_mex = data.query("iso_code == 'MEX'")

data_mex
```

```python
len(data_mex)
```

```python
data_mex.dropna()
```

```python
mex = pd.DataFrame({
    "date": data_mex["date"],
    "total_vaccinations": data_mex["total_vaccinations"],
    "people_vaccinated": data_mex["people_vaccinated"],
}).dropna()

mex
```

```python
import matplotlib.pyplot as plt

plt.bar(mex["date"][:30], mex["total_vaccinations"][:30])

plt.xticks(rotation=90)

plt.show()
```

```python
plt.bar(mex["date"][30:], mex["total_vaccinations"][30:])

plt.xticks(rotation=90)

plt.show()
```

```python
mex.query("date >= '2021-01-01' and date < '2021-02-01'")
```

```python
def is_date_odd(date):
    # TODO: Calcular si una fecha es fin de semana
    return int(date[-2:]) % 2 == 1

mex["date_is_odd"] = mex["date"].map(is_date_odd)

mex
```

```python
mex.query("date_is_odd")
```
