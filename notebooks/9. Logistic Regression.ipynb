# Regresión Logística

```python
import matplotlib.pyplot as plt

X = [
    [1, 1], # 3
    [1, 2], # 3
    [2, 1], # 3
    [4, 5], # 2
    [6, 7], # 2
    [2, 7], # 2
    [10, 3], # 1
    [13, 5], # 1
    [14, 4], # 1
    [11, 4], # 2
]

y = [3, 3, 3, 2, 2, 2, 1, 1, 1, 2]

n = len(X)

x1 = [X[i][0] for i in range(n)]
x2 = [X[i][1] for i in range(n)]

plt.scatter(x1, x2, c=y)
```

```python
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression()

reg.fit(X, y)

reg.score(X, y)
```

```python
import numpy as np

x_1 = np.linspace(0, 15, 100) # puntos sobre eje x1
x_2 = np.linspace(0, 8, 100) # puntos sobre eje x2

xx, yy = np.meshgrid(x_1, x_2) # matriz distribuida de puntos mezclados entre x1, x2

y_predict = reg.predict(np.c_[xx.ravel(), yy.ravel()]) # predice el valor de y, en cada x1, x2

# y es categórica

plt.scatter(xx, yy, c=y_predict)
plt.scatter(x1, x2, c=y)
```

```python
reg.predict([
    [4, 6], # 2
    [1, 3], # 3
    [10, 3], # 1
])
```

# Modelo Lineal Generalizado



## Regresión Poisson

```python
from sklearn.linear_model import PoissonRegressor

reg = PoissonRegressor()

reg.fit(X, y)

reg.score(X, y)
```

```python
x_1 = np.linspace(0, 15, 100) # puntos sobre eje x1
x_2 = np.linspace(0, 8, 100) # puntos sobre eje x2

xx, yy = np.meshgrid(x_1, x_2) # matriz distribuida de puntos mezclados entre x1, x2

y_predict = reg.predict(np.c_[xx.ravel(), yy.ravel()]) # predice el valor de y, en cada x1, x2

print(y_predict)

# y es categórica

plt.scatter(xx, yy, c=y_predict)
plt.scatter(x1, x2, c=y)
```
