## Histogramas y distribuciones

```python
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.exp(-(x - 5) ** 2)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

```python
plt.hist(y)
```

```python
x1 = np.array([1, 1.5, 1.2, 0.8, 0.3, 0.7, 1.1, 0.9, 0.8, 1, 1.05])

x1
```

```python
plt.hist(x1)
```

```python
plt.hist(x1, bins=5)
```

## Correlación

```python
x1 = np.array([1, 1.5, 1.2, 0.8, 0.3, 0.7, 1.1, 0.9, 0.8, 1, 1.05])
x2 = np.array([0.8, 1.3, 1.0, 0.6, 0.4, 0.5, 1.0, 0.75, 0.7, 0.8, 0.9])

plt.scatter(x1, x2)
```
