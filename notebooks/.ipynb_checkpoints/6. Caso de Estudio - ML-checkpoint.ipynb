## Adquisción

```python
import pandas as pd

data = pd.read_csv("datasets/agaricus-lepiota.data", header=None)

data
```

## Limpieza

1. cap-shape: bell=b,conical=c,convex=x,flat=f, knobbed=k,sunken=s
2. cap-surface: fibrous=f,grooves=g,scaly=y,smooth=s
3. cap-color: brown=n,buff=b,cinnamon=c,gray=g,green=r, pink=p,purple=u,red=e,white=w,yellow=y

```python
poisonous_map = {
    "p": 1,
    "e": 0
}

poisonous_logic_map = {
    "p": True,
    "e": False
}

cap_shape_map = {
    "b": 1,
    "c": 2,
    "f": 3,
    "x": 4,
    "k": 5,
    "s": 6
}

cap_shape_name_map = {
    "b": "bell",
    "c": "conical",
    "f": "flat",
    "x": "convex",
    "k": "knobbed",
    "s": "sunken"
}

cap_surface_map = {
    "f": 1,
    "g": 2,
    "y": 3,
    "s": 4
}

cap_surface_name_map = {
    "f": "fibrous",
    "g": "grooves",
    "y": "scaly",
    "s": "smooth"
}

cap_color_map = {
    "n": 1,
    "b": 2,
    "c": 3,
    "g": 4,
    "r": 5, 
    "p": 6,
    "u": 7,
    "e": 8,
    "w": 9,
    "y": 10
}

cap_color_name_map = {
    "n": "brown",
    "b": "buff",
    "c": "cinnamon",
    "g": "gray",
    "r": "green", 
    "p": "pink",
    "u": "purple",
    "e": "red",
    "w": "white",
    "y": "yellow"
}

data_study = pd.DataFrame({
    "CAP_SHAPE": data[1].map(cap_shape_map),
    "CAP_SHAPE_NAME": data[1].map(cap_shape_name_map),
    "CAP_SURFACE": data[2].map(cap_surface_map),
    "CAP_SURFACE_NAME": data[2].map(cap_surface_name_map),
    "CAP_COLOR": data[3].map(cap_color_map),
    "CAP_COLOR_NAME": data[3].map(cap_color_name_map),
    "POISONOUS": data[0].map(poisonous_map),
    "POISONOUS_LOGIC": data[0].map(poisonous_logic_map),
}) 

data_study
```

## Procesamiento

```python
X = pd.DataFrame({
    "CAP_SHAPE": data_study["CAP_SHAPE"],
    "CAP_SURFACE": data_study["CAP_SURFACE"],
    "CAP_COLOR": data_study["CAP_COLOR"],
}).to_numpy()

X
```

```python
Y = pd.DataFrame({
    "POISONOUS": data_study["POISONOUS"],
}).to_numpy()

Y.reshape(-1)
```

```python
size = len(X)
train_size = int(size * 0.8) # 80% del size (0.8 * size)
test_size = size - train_size

X_train = X[:train_size]
X_test = X[-test_size:]

Y_train = Y[:train_size]
Y_test = Y[-test_size:]
```

```python
! pip install scikit-learn
```

```python
from sklearn import svm

clf = svm.SVC()

clf.fit(X_train, Y_train.reshape(-1))
```

```python
fails = 0

for i in range(test_size):
    x = X_test[i]
    y = Y_test[i]
    
    yp = clf.predict([x])
    
    fails += abs(y[0] - yp[0])
    
    # print(x, yp, y)
    
fails, test_size, fails / test_size * 100
```
