```python
import pandas as pd

data = pd.read_csv("datasets/linear_regression_multi.csv")

data.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array(data["x1"])
x2 = np.array(data["x2"])
y = np.array(data["y"])

plt.scatter(x1, x2, c=y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.view_init(20, 60)

ax.scatter(x1, x2, y)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

X = np.array([
    x1, # [2.39, 0.018, ..., 1.18]
    x2, # [4.32, 6.28, ..., 1..99]
]).T

reg.fit(X, y)

# y = m1 * x1 + m2 * x2 + b
# y = 2.8 * x1 + (-1.05) * x2 + 6.69
print("coef (m1, m2)", reg.coef_)
print("intercept (b)", reg.intercept_)
print("R^2", reg.score(X, y))
```

```python
x1_m, x2_m = np.meshgrid(
    np.linspace(0, 10, 100),
    np.linspace(0, 10, 100),
)

X_m = np.array([
    x1_m.ravel(),
    x2_m.ravel()
]).T

# y_predict = m1 * x1_m + m2 * x2_m + b
# y_predict = reg.coef_[0] * x1_m + reg.coef_[1] * x2_m + reg.intercept_
y_predict = reg.predict(X_m)

plt.scatter(x1_m, x2_m, c=y_predict)
plt.scatter(x1, x2, c=y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.view_init(40, 60)

ax.scatter(x1, x2, y)
ax.scatter(x1_m, x2_m, y_predict)
```
