# Generador de datos para Regresiones Lineales

```python
import numpy as np

A = 0 # inicio
B = 10 # fin
N = 20 # mediciones
m = 3 # pendiente
b = 6 # intersecci√≥n
k = 1.5 # aleatoriedad

x = np.linspace(A, B, N)
y = m * (x + np.random.uniform(-1, 1, N) * k) + b
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)
```

```python
import pandas as pd

data = pd.DataFrame({
    "x": x,
    "y": y
})

data.to_csv("datasets/lineal_regresion.csv", index=False)
```
