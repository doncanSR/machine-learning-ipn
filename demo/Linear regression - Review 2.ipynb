```python
import pandas as pd
import numpy as np

data = pd.read_csv("datasets/linear_regression.csv")

data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt

plt.scatter(data["precio_1"], data["precio_2"])
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

X = np.array(data["precio_1"]).reshape((20, 1))

y = np.array(data["precio_2"])

reg.fit(X, y)

print("m", reg.coef_)
print("b", reg.intercept_)
print("R^2", reg.score(X, y))

y_predict = reg.predict(X)

plt.scatter(data["precio_1"], data["precio_2"])
plt.plot(data["precio_1"], y_predict, "red")
```
