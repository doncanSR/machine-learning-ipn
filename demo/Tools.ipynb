## Matplolib - Librería para la visualización de datos

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(subplot_kw={ "projection": "3d" })

X = np.linspace(-1, 1, 100)
Y = np.linspace(-1, 1, 200)

X, Y = np.meshgrid(X, Y)

Z = X ** 2 + Y ** 2

surf = ax.plot_surface(X, Y, Z)

plt.show()
```

## Pandas - Librería para el manejo de datasets

```python
import pandas as pd

edades = pd.Series([18, 25, 33, 56, 17, 29, 11, 54, 35])
generos = pd.Series(["M", "M", "H", "M", "H", "H", "H", "M", "M"])

data = pd.DataFrame({
    "EDAD": edades,
    "GENERO": generos
})

data["EDAD_CAT"] = data["EDAD"].map(lambda edad: int(edad / 10))
data["EDAD_LABEL"] = data["EDAD_CAT"].map({ 1: "INFANTE", 2: "ADOLESCENTE", 3: "JOVEN", 4: "ADULTO", 5: "MAYOR" })
data["GENERO_CAT"] = data["GENERO"].map({ "M": 1, "H": 2 })
data["GENERO_LABEL"] = data["GENERO_CAT"].map({ 1: "MUJER", 2: "HOMBRE" })

data
```

```python
import matplotlib.pyplot as plt

data["EDAD_LABEL"] = data["EDAD_CAT"].map({ 1: "Infante", 2: "Adolescente", 3: "Joven", 4: "Adulto", 5: "Mayor" })

gen_labels = list(set(data["GENERO_LABEL"]))
edades_sizes = [sum(data.query("GENERO_LABEL == '{}'".format(gen))["EDAD"]) for gen in gen_labels]

plt.bar(gen_labels, edades_sizes)

plt.show()
```
