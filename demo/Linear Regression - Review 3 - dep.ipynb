```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# y = m1 * x1 + m2 * x2 + b

data = pd.read_csv("datasets/linear_regression_multi.csv")

x1 = np.array(data["x1"])
x2 = np.array(data["x2"])
y = np.array(data["y"])

plt.scatter(x1, x2, c=y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.view_init(30, 60)

ax.scatter(x1, x2, y)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

X = np.array([
    x1,
    x2
]).T

reg.fit(X, y)

# y = m1 * x1 + m2 * x2 + b
print("coef (m1, m2)", reg.coef_)
print("intercept (b)", reg.intercept_)
print("R^2", reg.score(X, y))

x1_m, x2_m = np.meshgrid(np.linspace(0, 10, 100), np.linspace(0, 10, 100))

# y_predict = reg.coef_[0] * x1_m + reg.coef_[1] * x2_m + reg.intercept_
y_predict = reg.predict(np.array([
    x1_m.ravel(),
    x2_m.ravel()
]).T)

plt.scatter(x1_m, x2_m, c=y_predict)
plt.scatter(x1, x2, c=y)
```
