```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 20)
y = 5 * (x + np.random.uniform(-1, 1, 20) * 2) + 4

plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

X = x.reshape((20, 1))

reg.fit(X, y)

# y = m * x + b
print("coef (m)", reg.coef_)
print("intercept (b)", reg.intercept_)

reg.score(X, y) # R^2
```

```python
y_predict = [reg.coef_ * xi + reg.intercept_ for xi in x]

plt.scatter(x, y)
plt.plot(x, y_predict, "red")
```
