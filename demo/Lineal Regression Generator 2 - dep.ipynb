# Generador de datos para Regresiones Lineales - Parte II

```python
import numpy as np

A = 0 # inicio
B = 10 # fin
N = 20 # mediciones
m1 = 3 # pendiente
m2 = -1 # pendiente
b = 6 # intersecci√≥n
k1 = 1.5 # aleatoriedad
k2 = 2.5 # aleatoriedad

x1 = np.linspace(A, B, N)
x2 = np.linspace(A, B, N)

x1, x2 = np.meshgrid(x1, x2)

y = m1 * (x1 + np.random.uniform(-1, 1, N) * k1) + m2 * (x2 + np.random.uniform(-1, 1, N) * k2) + b + np.random.uniform(-1, 1)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x1, x2, c=y)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.view_init(20, 60)

plt.scatter(x1, x2, y)
```

```python
import pandas as pd

data = pd.DataFrame({
    "x1": x1,
    "x2": x2,
    "y": y
})

data.to_csv("datasets/linear_regression_multi.csv", index=False)
```
