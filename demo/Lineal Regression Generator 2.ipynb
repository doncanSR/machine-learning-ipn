# Generador de datos para Regresiones Lineales - Parte II

```python
import numpy as np

A = 0 # inicio
B = 10 # fin
N = 40 # mediciones
m1 = 3 # pendiente
m2 = -1 # pendiente
b = 6 # intersecci√≥n
k1 = 1.5 # aleatoriedad
k2 = 2.5 # aleatoriedad

points = []

for t in range(N):
    x1 = np.random.uniform(A, B)
    x2 = np.random.uniform(A, B)
    y = m1 * (x1 + np.random.uniform(-1, 1) * k1) + m2 * (x2 + np.random.uniform(-1, 1) * k2) + b + np.random.uniform(-1, 1)
    points.append((x1, x2, y))
```

```python
import matplotlib.pyplot as plt

x1, x2, y = zip(*points)

plt.scatter(x1, x2, c=y)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.view_init(20, 60)

plt.scatter(x1, x2, y)
```

```python
import pandas as pd

data = pd.DataFrame({
    "x1": x1,
    "x2": x2,
    "y": y
})

data.to_csv("datasets/linear_regression_multi.csv", index=False)
```
