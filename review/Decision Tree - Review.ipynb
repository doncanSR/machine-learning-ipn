```python
# Muestras -> dataset (Aprendizaje Supervisado)

# Características
X = [
    [1, 2, 3], # 2
    [4, 5, 7], # 1
    [2, 6, 9], # 2
    [8, 7, 4], # 3
    [1, 3, 2], # 1
    [4, 5, 6], # 1
]

# Clases (etiquetas)
y = [2, 1, 2, 3, 1, 1]
```

```python
# Árbol de decisión

from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()

clf.fit(X, y)

clf.predict([
    [1, 2, 3], # 2
    [1.1, 1.9, 2.8], # 2
    [4, 4, 4], # 1
    [4, 7, 4] # 1
])
```

```python
from sklearn.tree import plot_tree

plot_tree(clf)
```

```python
clf.score(X, y)
```
